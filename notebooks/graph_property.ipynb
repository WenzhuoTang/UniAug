```python
def plot_property(base_dataset, x='density', y='entropy', alpha_base=0.5, alpha=0.8):
    plt.figure(figsize=(10, 6))

    x_scaler = base_dataset.scaler_dict[x]
    y_scaler = base_dataset.scaler_dict[y]
    y_prop = y_scaler.transform(base_dataset.property_dict[y])

    plt.scatter(
        x_scaler.transform(base_dataset.property_dict[x]),
        y_scaler.transform(base_dataset.property_dict[y]), 
        label='pretraining', s=2, alpha=alpha_base
    )

    plt.scatter(
        x_scaler.transform(cora.property_dict[x]),
        y_scaler.transform(cora.property_dict[y]), 
        color='red', marker='o', s=30, alpha=alpha, label='cora'
    )
    plt.scatter(
        x_scaler.transform(citeseer.property_dict[x]),
        y_scaler.transform(citeseer.property_dict[y]), 
        color='red', marker='v', s=30, alpha=alpha, label='citeseer'
    )
    plt.scatter(
        x_scaler.transform(pubmed.property_dict[x]),
        y_scaler.transform(pubmed.property_dict[y]), 
        color='red', marker='^', s=30, alpha=alpha, label='pubmed'
    )
    # plt.scatter(
    #     x_scaler.transform(ddi.property_dict[x]),
    #     y_scaler.transform(ddi.property_dict[y]), 
    #     color='red', marker='*', s=30, alpha=alpha, label='ddi'
    # )
    plt.scatter(
        x_scaler.transform(collab.property_dict[x]),
        y_scaler.transform(collab.property_dict[y]), 
        color='red', marker='X', s=30, alpha=alpha, label='collab'
    )

    plt.scatter(
        x_scaler.transform(power.property_dict[x]),
        y_scaler.transform(power.property_dict[y]), 
        color='black', marker='o', s=30, alpha=alpha, label='power'
    )

    plt.scatter(
        x_scaler.transform(yst.property_dict[x]),
        y_scaler.transform(yst.property_dict[y]), 
        color='black', marker='v', s=30, alpha=alpha, label='yst'
    )

    plt.scatter(
        x_scaler.transform(erd.property_dict[x]),
        y_scaler.transform(erd.property_dict[y]), 
        color='black', marker='*', s=30, alpha=alpha, label='erd'
    )

    plt.scatter(
        x_scaler.transform(amazon_photo.property_dict[x]),
        y_scaler.transform(amazon_photo.property_dict[y]), 
        color='black', marker='^', s=30, alpha=alpha, label='amazon_photo'
    )

    plt.scatter(
        x_scaler.transform(flickr.property_dict[x]),
        y_scaler.transform(flickr.property_dict[y]), 
        color='black', marker='X', s=30, alpha=alpha, label='flickr'
    )

    perm1 = np.random.permutation(range(len(ogbg_molhiv)))
    plt.scatter(
        x_scaler.transform(ogbg_molhiv.property_dict[x][perm1[:200]]),
        y_scaler.transform(ogbg_molhiv.property_dict[y][perm1[:200]]), 
        color='gray', marker='o', s=2, alpha=0.8, label='ogbg-molhiv'
    )

    perm2 = np.random.permutation(range(len(ogbg_ppa)))
    plt.scatter(
        x_scaler.transform(ogbg_ppa.property_dict[x][perm2[:200]]),
        y_scaler.transform(ogbg_ppa.property_dict[y][perm2[:200]]), 
        color='peru', marker='o', s=2, alpha=0.6, label='ogbg-ppa'
    )

    perm3 = np.random.permutation(range(len(ogbg_code2)))
    plt.scatter(
        x_scaler.transform(ogbg_code2.property_dict[x][perm3[:200]]),
        y_scaler.transform(ogbg_code2.property_dict[y][perm3[:200]]), 
        color='darkviolet', marker='o', s=2, alpha=0.3, label='ogbg-code2'
    )

    # plt.scatter(
    #     x_scaler.transform(cora_seg.property_dict[x]),
    #     y_scaler.transform(cora_seg.property_dict[y]), 
    #     color='green', marker='o', s=20, alpha=alpha, label='cora_seg'
    # )

    # plt.scatter(
    #     x_scaler.transform(citeseer_seg.property_dict[x]),
    #     y_scaler.transform(citeseer_seg.property_dict[y]), 
    #     color='green', marker='v', s=20, alpha=alpha, label='citeseer_seg'
    # )
    
    # plt.scatter(
    #     x_scaler.transform(pubmed_seg.property_dict[x]),
    #     y_scaler.transform(pubmed_seg.property_dict[y]), 
    #     color='green', marker='^', s=20, alpha=alpha, label='pubmed_seg'
    # )
    # plt.scatter(
    #     x_scaler.transform(ddi_seg.property_dict[x]),
    #     y_scaler.transform(ddi_seg.property_dict[y]), 
    #     color='green', marker='*', s=20, alpha=alpha, label='ddi_seg'
    # )
    # plt.scatter(
    #     x_scaler.transform(collab_seg.property_dict[x]),
    #     y_scaler.transform(collab_seg.property_dict[y]), 
    #     color='green', marker='X', s=20, alpha=alpha, label='collab_seg'
    # )

    # Adding labels and title
    plt.xlabel(x)
    plt.ylabel(y)
    plt.title('Structural Properties (normalized)')
    plt.legend() 
    
    # Display the plot
    plt.show()
```
