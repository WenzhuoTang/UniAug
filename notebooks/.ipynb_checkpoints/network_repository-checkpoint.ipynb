```python
postfixes = ['mtx', 'txt', 'txt.gz', 'edges']

def read_files(root, data_name):
    files = os.listdir(root)
    # data_file = [x for x in files if x.startswith(data_name) and not x.endswith('zip')]
    data_file = [x for x in files if any([x.endswith(pf) for pf in postfixes]) and not x.startswith('readme')]
    if len(data_file) > 0:
        if len(data_file) > 1:
            print('######### data files more than 1 #########')
            print(f'######### data_name: {data_name} #########')
            print(data_file)
        data_file = data_file[0]
    
        flag = True
        try:
            edges_coo = mmread(f'{root}/{data_file}')
            edge_index = torch.stack(
                (torch.from_numpy(edges_coo.col), torch.from_numpy(edges_coo.row))
            ).long()
            edge_attr = torch.from_numpy(edges_coo.data)
    
        except:
    
            try:
                edges_df = pd.read_csv(f'{root}/{data_file}', header=None, sep=' ')
                if data_file.endswith('mtx'):
                    # remove irrelavent rows
                    counter = 0
                    while True:
                        try:
                            int(edges_df.iloc[counter, 0])
                            break
                        except:
                            counter += 1
                    edges_df = edges_df.iloc[counter:]
                    edges_df.iloc[:,0] = edges_df.iloc[:,0].astype(int)
                    edges_df.iloc[:,1] = edges_df.iloc[:,1].astype(int)
        
                elif data_file.endswith('txt.gz'):
                    # remove irrelavent rows
                    counter = 0
                    while True:
                        try:
                            int(str(edges_df.iloc[counter, 0]).split('\t')[0])
                            break
                        except:
                            counter += 1
                    temp_df = edges_df.iloc[counter:, 0].copy()
                    edges_df = temp_df.str.split('\t', n=1, expand=True)
        
                edge_index = torch.stack((
                    torch.tensor(edges_df.iloc[:,0].values.astype(int), dtype=torch.long), 
                    torch.tensor(edges_df.iloc[:,1].values.astype(int), dtype=torch.long)
                ))
                if edges_df.shape[1] > 2 and not any(edges_df.iloc[:,2].isna()):
                    edge_attr = torch.tensor(edges_df.iloc[:,2].values.astype(float))
                else:
                    edge_attr = None
    
            except:
                edge_index = None
                edge_attr = None
                flag = False

    else:
        print('######### data files empty #########')
        print(f'######### data_name: {data_name} #########')
        print(files)
        edge_index = None
        edge_attr = None
        flag = False

    if edge_index is not None and edge_index.shape[1] == 0:
        flag = False

    if flag and edge_attr is None:
        edge_attr = torch.ones(edge_index.shape[1])

    return edge_index, edge_attr, flag
```

```python
from torch_geometric.utils import to_undirected, contains_isolated_nodes, remove_isolated_nodes

data_names = []
data_list = []
submeta['accessibility'] = True

for name in tqdm(submeta.index):
    data_root = f'{root}/data/{name}'
    edge_index, edge_attr, flag = read_files(data_root, name)
    data_flag = False
    if flag:
        if contains_isolated_nodes(edge_index):
            edge_index, edge_attr, _ = remove_isolated_nodes(edge_index, edge_attr)

        if edge_index.numel() > 0:
            data_names.append(name)
            edge_index, edge_attr = to_undirected(edge_index, edge_attr)
            data = Data(
                edge_index=edge_index, edge_attr=edge_attr, num_nodes=edge_index.max() + 1
            )
            data_list.append(data)
            data_flag = True
    submeta.loc[name, 'accessibility'] = data_flag
```

```python
def plot_property_with_cluster(property_dict, scaler_dict, base_labels, x='density', y='entropy', alpha_base=0.5, 
                               plot_downstream=True, alpha=0.8):
    plt.figure(figsize=(10, 6))

    x_scaler = scaler_dict[x]
    y_scaler = scaler_dict[y]

    for k in np.unique(base_labels):
        idx = np.where(base_labels == k)[0]
        plt.scatter(
            x_scaler.transform(property_dict[x][idx]),
            y_scaler.transform(property_dict[y][idx]), 
            label=f'cluster_{k}', s=10, alpha=alpha_base
        )

    if plot_downstream:
        plt.scatter(
            x_scaler.transform(cora.property_dict[x]),
            y_scaler.transform(cora.property_dict[y]), 
            color='black', marker='o', s=30, alpha=alpha, label='cora'
        )
        plt.scatter(
            x_scaler.transform(citeseer.property_dict[x]),
            y_scaler.transform(citeseer.property_dict[y]), 
            color='black', marker='v', s=30, alpha=alpha, label='citeseer'
        )
        plt.scatter(
            x_scaler.transform(pubmed.property_dict[x]),
            y_scaler.transform(pubmed.property_dict[y]), 
            color='black', marker='^', s=30, alpha=alpha, label='pubmed'
        )
        plt.scatter(
            x_scaler.transform(ogbl_collab.property_dict[x]),
            y_scaler.transform(ogbl_collab.property_dict[y]), 
            color='black', marker='X', s=30, alpha=alpha, label='ogbl-collab'
        )

        perm1 = np.random.permutation(range(len(enzymes)))
        plt.scatter(
            x_scaler.transform(enzymes.property_dict[x][perm1[:200]]),
            y_scaler.transform(enzymes.property_dict[y][perm1[:200]]), 
            color='indigo', marker='o', s=2, alpha=0.8, label='enzymes'
        )
    
        perm2 = np.random.permutation(range(len(reddit5k)))
        plt.scatter(
            x_scaler.transform(reddit5k.property_dict[x][perm2[:200]]),
            y_scaler.transform(reddit5k.property_dict[y][perm2[:200]]), 
            color='peru', marker='o', s=2, alpha=0.6, label='reddit5k'
        )
    
        perm3 = np.random.permutation(range(len(github)))
        plt.scatter(
            x_scaler.transform(github.property_dict[x][perm3[:200]]),
            y_scaler.transform(github.property_dict[y][perm3[:200]]), 
            color='lime', marker='o', s=2, alpha=0.6, label='github'
        )
    
        perm4 = np.random.permutation(range(len(dd)))
        plt.scatter(
            x_scaler.transform(dd.property_dict[x][perm4[:200]]),
            y_scaler.transform(dd.property_dict[y][perm4[:200]]), 
            color='aqua', marker='o', s=2, alpha=0.6, label='dd'
        )
    
        perm5 = np.random.permutation(range(len(proteins)))
        plt.scatter(
            x_scaler.transform(proteins.property_dict[x][perm5[:200]]),
            y_scaler.transform(proteins.property_dict[y][perm5[:200]]), 
            color='violet', marker='o', s=2, alpha=0.6, label='proteins'
        )

    # Adding labels and title
    plt.xlabel(x)
    plt.ylabel(y)
    plt.title('Structural Properties (normalized)')
    plt.legend() 
    
    # Display the plot
    plt.show()
```
